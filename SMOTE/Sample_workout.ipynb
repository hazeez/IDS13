```python
def age_bin(age):
    if 0 < age[0] <= 10:
        return '1-10'
    elif 10 < age[0] <= 20:
        return '11-20'
    elif 21 < age[0] <= 30:
        return '21-30'
    elif 31 < age[0] <= 40:
        return '31-40'
    elif 41 < age[0] <= 50:
        return '41-50'
    else:
        return 'Greater than 50'
    
```

```python
data['Age_bin'] = data[['Age']].apply(age_bin,axis=1)
```

```python
def preg_bin(preg):
    if 0 <= preg[0] <= 3:
        return '0-3'
    elif 4 <= preg[0] <= 6:
        return '4-6'
    elif 7 <= preg[0] <= 10:
        return '7-10'
    else:
        return 'Greater than 10'
```

```python
data['Preg_bin'] = data[['Pregnancies']].apply(preg_bin,axis=1)
```

```python
model_df = pd.get_dummies(data.drop(['Pregnancies', 'Age'],axis=1))
```

```python
X = model_df.drop('Outcome', axis=1)
y = model_df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scx = StandardScaler()
model_df_std = scx.fit_transform(model_df.drop('Outcome', axis=1))
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, y)
np.set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)

print(features[0:5,:])
print(test.get_support(indices=False))

```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
rfe = RFE(model, 5)
fit = rfe.fit(X, y)
print("Num Features: ", fit.n_features_)
print("Selected Features: ", fit.support_)
print("Feature Ranking: ", fit.ranking_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
result = model.score(X_test, y_test)
print("Accuracy: ", (result*100.0))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv=5)
scores
```

```python
from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
cross_val_score(model, X, y, cv=cv)
```

```python
print(f'{scores.mean()} ({scores.std()})')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DT', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
results = []
names = []
for name, model in models:
    cv_results = cross_val_score(model, X, y, cv=10, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    print(f'{name}: {cv_results.mean()} ({cv_results.std()})')
```

```python
sns.boxplot(x=names, y=results)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.pipeline import FeatureUnion
from sklearn.decomposition import PCA
```

```python
features = []
features.append(('pca', PCA(n_components=3)))
features.append(('select_best', SelectKBest(k=6)))
feature_union = FeatureUnion(features)

estimators = []
estimators.append(('feature_union', feature_union))
estimators.append(('logistic', LogisticRegression()))
model = Pipeline(estimators)
results = cross_val_score(model, X, y, cv=10)
print(results.mean())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
estimators = []
model1 = LogisticRegression()
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('decisiontree', model2))
model3 = SVC()
estimators.append(('svm', model3))
ensemble = VotingClassifier(estimators)
results = cross_val_score(ensemble, X, y, cv=10)
print(results.mean())
```

```python
from sklearn.ensemble import StackingClassifier
```

```python
stack_clf = StackingClassifier(estimators, final_estimator=RandomForestClassifier(), cv=5)
```

```python
stack_clf.fit(X_train, y_train)
```

```python
cross_val_score(stack_clf, X, y, cv=5).mean()
```

```python
rf = RandomForestClassifier()
params = {'n_estimators':[1000, 2000, 5000, 7000], 'max_depth':[5,10,15,20]}
from sklearn.model_selection import RandomizedSearchCV
```

```python
rsearch = RandomizedSearchCV(estimator=rf, param_distributions=params,
n_iter=2, cv=2, random_state=101)
rsearch.fit(X, y)
print(rsearch.best_score_)
print(rsearch.best_estimator_)
```
